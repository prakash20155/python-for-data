```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np

import datetime
%matplotlib inline
```

```python
start=datetime.datetime(2006,1,1)
end=datetime.datetime(2016,1,1)
```

```python
# Bank of America
BAC =data.DataReader("BAC",'yahoo',start,end)  # look in solution  sometimes google dont work
# CitiGroup
C=data.DataReader('C','yahoo',start,end)
#Goldman Sachs
GS=data.DataReader('GS','yahoo',start,end)
# JPMorgan Chase
JPM = data.DataReader("JPM", 'yahoo', start, end)

# Morgan Stanley
MS = data.DataReader("MS", 'yahoo', start, end)

# Wells Fargo
WFC = data.DataReader("WFC", 'yahoo', start, end)
```

```python
BAC
```

```python
tickers=['BAC','C','GS','JPM','MS','WFC']
```

```python
bank_stocks=pd.concat([BAC,C,GS,JPM,MS,WFC],axis=1,keys=tickers)
```

```python
bank_stocks.head(5)
```

```python
bank_stocks.columns.names=['Bank Ticker','Stock Info']
```

```python
bank_stocks.head()
```

```python
# Exploratory data analysis(EDA)
```

```python
# max close price for each bank stock
for tick in tickers:
    print(tick,bank_stocks[tick]['Close'].max())
```

```python
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()     # xs is crosssection
```

```python
# empty dataframe called return

returns=pd.DataFrame()
```

```python
for tick in tickers:
    returns[tick+'Return']=bank_stocks[tick]['Close'].pct_change()
```
