```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
# functional method
plt.plot(x,y)
```

```python
plt.plot(x,y,'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)          # row,col,plotno.
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# oop method
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])       # distance from  left,bottom, width, height
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])       
axes2=fig.add_axes([0.4,0.3,0.2,0.5])
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])      
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger plot')

axes2.plot(y,x)
axes2.set_title('Smaller plot')
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2)
ax.plot(x,x**3)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X squared')
ax.plot(x,x**3, label='Y cubed')

ax.legend()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X squared')
ax.plot(x,x**3, label='Y cubed')

ax.legend(loc=10)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X squared')
ax.plot(x,x**3, label='Y cubed')

ax.legend(loc=(0.1,0.1))
```

```python
# plot appearance
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',linewidth=10,alpha=0.5)     # alpha for transparency  and can type lw for linewidth
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',lw=5,ls='--')       # ls = linestyle
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',linewidth=3,ls='-',marker='o',markersize=10,markerfacecolor='yellow',markeredgewidth=20,markeredgecolor='green')            #marker marks the value of x vs y 
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',lw=1,ls='--') 

ax.set_xlim([0,1])                # x limit (upperbound,lowerbound)
ax.set_ylim([0,2])
```
